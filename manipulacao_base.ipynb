```python
import re
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_excel('D:/papers/direito/scrapping/base.xlsx')
df1
```

```python
regex = '([Nn][ãa]o |[Dd]es|[Ii]m|[Nn]egar )(provido|conhecido|provimento)(s)?' # definindo o regex para capturar casos de não provimento
df1['nao_provimento'] = np.where(df1.ementa.str.contains(regex), 1, 0) # caso haja não provimento na ementa, atribuimos 1 à dummy

df1['ministro_relator'], df1['ministro_relator_acordao'] = df1['ministro_relator'].str.split('\spara\sAcórdão', 1).str # separando os ministros entre relatores e relatores do acórdão
df1.loc[df1['ministro_relator_acordao'].isnull(),'ministro_relator_acordao'] = df1['ministro_relator'] # caso não haja um relator para o acórdão, este será o relator "original"

df1.to_excel('D:/papers/direito/scrapping/base2.xlsx', index=False) # salvando a base de dados
df1
```
